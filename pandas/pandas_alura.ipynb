*explicação*

Count: Conta o número de valores não nulos em uma coluna, útil para detectar dados faltantes.


Mean: Média aritmética dos valores, mas pode ser distorcida por valores extremos.


Std: Desvio padrão, mede a dispersão dos dados em relação à média.


Min e Max: Valores mínimo e máximo da amostra, indicando a amplitude dos dados.


Quartis: Valores que dividem os dados em partes iguais, como a mediana (50%), que indica o valor que separa metade dos dados.


**Análise exploratória de dados**

Qual o valor médio de aluguél por tipo de imóvel?

```python
alugueis.head()
```

```python
#calcula a média de todos os valores e arredonda
round(alugueis['Valor'].mean())
```

```python
#agroupa pela coluna 'Tipo' e retorna a média de todos os dados numéricos
alugueis.groupby('Tipo').mean(numeric_only=True)
```

```python
#calcula média de valor de cada tipo de imóvel, ordenando pelo valor
alugueis.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
#atribui a uma variável 
df_preco_tipo = alugueis.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

#exibe um gráfico de tipo barh (barras horizontais), com tamanho e cor determinados
df_preco_tipo.plot(kind='barh', figsize=(12, 6), color='red')
```

**Removendo os imóveis comerciais**

Não queremos imóveis comerciais no nosso dataset devido ao seu alto preço

```python
#retorna todo único nome de tipo de imóvel
alugueis.Tipo.unique()
```

```python
#serparando os imóveis comerciais
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
                                            
```

```python
#retorna todos os imóveis comerciais 
alugueis.query('@imoveis_comerciais in Tipo')
```

```python
#atribui a uma variável os imóveis não comerciais
nao_comerciais = alugueis.query('@imoveis_comerciais not in Tipo')
nao_comerciais.head()
```

```python
#retorna todos os tipos de dados, que agora não possuem imóveis comerciais
nao_comerciais.Tipo.unique()
```

```python
#atribui a uma variável o novo dataset
df_preco_tipo = nao_comerciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

#exibe um gráfico de tipo barh (barras horizontais), com tamanho e cor determinados
df_preco_tipo.plot(kind='barh', figsize=(12, 6), color='red')
```

**Qual o percentual de cada imóvel em nossa base de dados?**

```python
#retorna todos os tipos únicos de imóveis
nao_comerciais.Tipo.unique()
```

```python
#conta os valores de cada tipo de imóvel e retorna o valor percentual
nao_comerciais.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')
```

```python
from turtle import color

#atribui o dataframe a uma variável
nao_comerciais_percentual_tipo = nao_comerciais.Tipo.value_counts(normalize=True).to_frame().sort_values('proportion')

#exibe um gráfico de barras, com tamanho determinado, cor vermelha, e rotulando as coordenadas X e Y
nao_comerciais_percentual_tipo.plot(kind='bar', figsize=(12, 5), color='red', xlabel='Tipos', ylabel='Porcentagem')
```
