**desafio:**
1. Verifique se o arquivo CSV está separado por vírgula ou ponto e vírgula.
2. A codificação do arquivo é ISO-8859-1.
3. As três primeiras linhas linhas do arquivo podem ser desconsideradas, pois o cabeçalho só começa na quarta linha.
4. As 9 últimas linhas também podem ser desconsideradas, pois são apenas informações sobre onde os dados foram obtidos.
5. Para deletar as últimas linhas é necessário adicionar o parâmetro engine='python'.

```python
#lê o link com os arquivos csv, com separação por ';', codificação ISO8859, e pulando as primeiras e ultimas linhas
dados_hospitais = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv', sep=';', encoding='ISO-8859-1',skiprows=3, skipfooter=9)
dados_hospitais
```

**Manipulando arquivos JSON**

Lendo arquivos JSON

```python
#Faz a leitura do arquivo JSON
dados_pacientes = pd.read_json('pacientes.json')
```

```python
#retorna o cabeçalho do arquivo
dados_pacientes.head()
```

```python
#atribui a uma variável a leitura do arquivo json e retorna o cabeçalho
dados_pacientes_2 = pd.read_json('pacientes_2.json')
dados_pacientes_2.head()
```

```python
#normaliza o dataframe, separando os dados da tabela 'pacientes'
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
```

```python
#imprime o cabeçalho do df
df_normalizado.head()
```

```python
#salva o df normalizado em um arquivo json
df_normalizado.to_json('historico_pacientes_normalizado.json')
```

```python
#lê o novo arquivo salvo
pacientes_normalizados = pd.read_json('historico_pacientes_normalizado.json')
```

```python
#imprime o cabeçalho do df
pacientes_normalizados.head()
```

Obtendo arquivos JSON de APIs

```python
#importa a biblioteca JSON e requests
import json
import requests
```

```python
#pega os dados da API em formato json
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
```

```python
#atribui a uma variável o carregamento dos dados em formato json
resultado = json.loads(dados_frutas.text)
```

```python
#atribui a uma variável a leitura dos dados
frutas = pd.DataFrame(resultado)
```

```python
#imprime o cabeçalho do df
frutas.head()
```

Desafio: normalize o dataframe

```python
#normalizando o df separand os dados da coluna 'nutritions'
frutas_normalizadas = pd.json_normalize(resultado)
```

```python
#imprime o cabeçalho do df
frutas_normalizadas.head()
```
